```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
with open('translated_output.txt', 'r',encoding="utf-8") as file:
    text = file.read()
text_splitter=RecursiveCharacterTextSplitter(
    chunk_size=200,
    chunk_overlap=50,
    length_function=len,
    add_start_index=True,
)
chunks=text_splitter.split_text(text)
print(f"Number of chunks: {len(chunks)}")
print(chunks[2])
```

```python
pip install sentence-transformers
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('all-mpnet-base-v2')
embeddings = model.encode(chunks, show_progress_bar=True)
```

```python
print("First embedding vector:")
print(embeddings[0])
```

```python
print("Embedding vector length:", len(embeddings[0]))  # Output: 768
```

```python
import faiss
import numpy as np
import pickle

# Let's say you already have your chunk embeddings
# and the corresponding text chunks
# embeddings: List of vectors (np.array)
# chunks: List of original chunk texts

# Convert embeddings to numpy array
embedding_matrix = np.array(embeddings).astype("float32")

# Initialize FAISS index
dimension = embedding_matrix.shape[1]  # e.g., 768
index = faiss.IndexFlatL2(dimension)

# Add embeddings to FAISS
index.add(embedding_matrix)

# Optional: Save index + chunks for reuse
faiss.write_index(index, "faiss_index.idx")
with open("chunks.pkl", "wb") as f:
    pickle.dump(chunks, f)
```

```python
# Load index and chunks
index = faiss.read_index("faiss_index.idx")
with open("chunks.pkl", "rb") as f:
    chunks = pickle.load(f)

# Embed the user's question
question = "What are the two options discussed for making ₹5 lakhs — part-time with a team or full-time individually? Which seems easier and why?"
question_embedding = model.encode([question]).astype("float32")

# Search FAISS for top-k relevant chunks
D, I = index.search(question_embedding, k=3)

# Get top 3 matching chunks
top_chunks = [chunks[i] for i in I[0]]
print("Top relevant chunks:\n", "\n".join(top_chunks))
```
